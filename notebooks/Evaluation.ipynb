```python
def plot_metrics_14classes(df, code):
    plt.figure()
    model1 = plt.scatter(df.train_accuracy_eval,
                         df.name_code,
                         c='magenta',
                         label='train')
    model2 = plt.scatter(df.test_accuracy_eval,
                         df.name_code,
                         c='cyan',
                         label='test')
    model3 = plt.scatter(df.valid_accuracy_eval,
                         df.name_code,
                         c='gray',
                         label='valid')

    plt.xlabel('Accuracy')
    plt.ylabel('Models')
    plt.title("Accuracy of models trained for whole MURA dataset (14 classes)")
    plt.legend()
    major_ticks = np.arange(10, 91, 5)
    plt.xticks(major_ticks)    
    plt.grid(alpha=0.5, linestyle='--')
    plt.savefig('data/plots/evaluation_accuracy_one_model_14classes' +  code + '.png')
    plt.show()
```

```python
finetuned_each_part = finetuned[finetuned.part_body == 'XR_SHOULDER']
```

```python
def plot_each_part(bodypart, df, code):
    plt.figure()    
    model1 = plt.scatter(df[df.part_body == bodypart].train_accuracy_eval,
                         df[df.part_body == bodypart].name_code,
                         c='magenta',
                         label='train')
    model2 = plt.scatter(df[df.part_body == bodypart].test_accuracy_eval,
                         df[df.part_body == bodypart].name_code,
                         c='cyan',
                         label='test')
    model3 = plt.scatter(df[df.part_body == bodypart].valid_accuracy_eval,
                         df[df.part_body == bodypart].name_code,
                         c='gray',
                         label='valid')
    major_ticks = np.arange(50, 95, 5)
    plt.xticks(major_ticks)
    plt.xlabel('Accuracy')
    plt.ylabel('Models')
    plt.grid(alpha=0.5, linestyle='--')
    plt.title("Accuracy of models trained for " + bodypart + code + " (2 classes)")
    plt.savefig('data/plots/evaluation_accuracy_' + bodypart+code + '.png')
    plt.legend()
    plt.show()
```

```python
bodyparts = ['XR_ELBOW',
 'XR_FINGER',
 'XR_FOREARM',
 'XR_HAND',
 'XR_HUMERUS',
 'XR_SHOULDER',
 'XR_WRIST']
```

```python
for bodypart in bodyparts:
    plot_each_part(bodypart, results_each_part, '')
```
